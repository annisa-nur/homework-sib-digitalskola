```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load data
data = pd.read_csv('/content/Sales.txt', sep="\s+", header=None, names=['Sales', 'Advertising'])

# Split data into X and y
X = data[['Sales']]
y = data['Advertising']

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predictions
sales_to_predict = np.array([[50], [100], [150]])
predictions = model.predict(sales_to_predict)

# Evaluate the model
y_pred = model.predict(X)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

# Print results
print(f"Predicted advertising costs for sales 50, 100, 150: {predictions}")
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")

# Plot results
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Sales')
plt.ylabel('Advertising')
plt.title('Sales vs Advertising')
plt.show()
```
